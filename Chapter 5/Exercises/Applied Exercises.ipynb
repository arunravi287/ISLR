```python
from sklearn.model_selection import train_test_split

default_train, default_validation = train_test_split(default, test_size=0.5, random_state=42)

X = default[["income","balance"]]
X_train = default_train[["income","balance"]]
X_validation = default_validation[["income","balance"]]

Y = default["default"]
Y_train = default_train["default"]
Y_validation = default_validation["default"]
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,accuracy_score

#performance on training data
predictions=lr.predict(X_train)
print(pd.DataFrame(confusion_matrix(Y_train,predictions),['True No','True Yes'], ['Predicted No','Predicted Yes']))
print("Precision: " + str(precision_score(Y_train,predictions,average="macro")))
print("Recall: " + str(recall_score(Y_train,predictions,average="macro")))
print("Accuracy: " + str(accuracy_score(Y_train,predictions,normalize=True)))
print("F1: " +str(f1_score(Y_train,predictions,average="macro")))
```

```python
#performance on validation set
predictions=lr.predict(X_validation)
print(pd.DataFrame(confusion_matrix(Y_validation,predictions),['True No','True Yes'], ['Predicted No','Predicted Yes']))
print("Precision: " + str(precision_score(Y_validation,predictions,average="macro")))
print("Recall: " + str(recall_score(Y_validation,predictions,average="macro")))
print("Accuracy: " + str(accuracy_score(Y_validation,predictions,normalize=True)))
print("F1: " +str(f1_score(Y_validation,predictions,average="macro")))
```

```python
def getParameters(X,Y):
    lr=LogisticRegression()
    lr.fit(X,Y)
    return lr.coef_
```

```python
from sklearn.utils import resample

parameters=np.zeros(shape=(1000,2))
for count in range(1,1000):
    default_resampled = resample(default,n_samples=default.shape[0])
    X_resampled = default_resampled[["income","balance"]]
    Y_resampled = default_resampled["coded"]
    parameters[count-1] = getParameters(X_resampled,Y_resampled)
    
print("Mean of Coefficient for income: " + str(parameters[:,0].mean()))
print("Mean of Coefficient for balance: " + str(parameters[:,1].mean()))
print("Standard Deviation of Coefficient for income: " + str(parameters[:,0].std()))
print("Standard Deviation of Coefficient for balance: " + str(parameters[:,1].std()))
```

```python
#7a -> Logistic Regression using the whole Data Set
X=weekly[["Lag1","Lag2"]]
Y=weekly["Direction"]

lr=LogisticRegression()
lr.fit(X,Y)

predictions=lr.predict(X)
print(pd.DataFrame(confusion_matrix(Y,predictions),['True Down','True Up'], ['Predicted Down','Predicted Up']))
print("Precision: " + str(precision_score(Y,predictions,average="macro",zero_division=0)))
print("Recall: " + str(recall_score(Y,predictions,average="macro")))
print("Accuracy: " + str(accuracy_score(Y,predictions,normalize=True)))
print("F1: " +str(f1_score(Y,predictions,average="macro")))
```

```python
#7b -> Logistic Regression using the whole Data Set but the first observation
def loocv(data,n):
    return pd.concat([data[0:n],data[n+1:data.shape[0]]])
```

```python
X=weekly[["Lag1","Lag2"]]
Y=weekly["Direction"]

X_modified = loocv(X,0)
Y_modified = loocv(Y,0)

lr=LogisticRegression()
lr.fit(X_modified,Y_modified)
```

```python
#7c -> predict the first observation
prediction=lr.predict(X.iloc[[0]])
prediction[0]
```

This observation is not correctly classified.

```python
# d -> LOOCV
errors=np.zeros(shape=(weekly.shape[0],1))
for i in range(1,weekly.shape[0]):
    modified = loocv(weekly,i-1)
    X=modified[["Lag1","Lag2"]]
    Y=modified["Direction"]
    
    lr.fit(X,Y)
    errors[i-1] = 1 if lr.predict(X.iloc[[i-1]])!=Y.iloc[i-1] else 0
```

```python
#7e -> find avarage error
errors.mean()
```

### Question 8

```python
#8a -> generate a simulated data set
np.random.seed(42)
X=np.random.normal(0,1,100)
Y=np.random.normal(0,1,100)
Y=X-2*np.power(X,2)+np.random.normal(0,1,100)
```

$n$ = 100<br>
$p$ = $X$ and $X^2$

```python
#8b -> generate a scatter plot
sns.scatterplot(x=X,y=Y)
plt.show()
```

```python
simulated=pd.DataFrame({"X": X, "X2": np.power(X,2), "X3":  np.power(X,3), "X4": np.power(X,4), "Y": Y})
simulated.head()
```

```python
from sklearn.model_selection import LeaveOneOut,cross_val_score
from sklearn.linear_model import LinearRegression

#8c -> LOOCV
loo=LeaveOneOut()
lr=LinearRegression()

scores = cross_val_score(lr,simulated["X"].to_numpy().reshape(-1,1),simulated["Y"],cv=loo,scoring='neg_mean_squared_error')
print(scores.mean()*(-1))
```

```python
scores = cross_val_score(lr,simulated[["X","X2"]],simulated["Y"],cv=loo,scoring='neg_mean_squared_error')
print(scores.mean()*(-1))
```

```python
scores = cross_val_score(lr,simulated[["X","X2","X3"]],simulated["Y"],cv=loo,scoring='neg_mean_squared_error')
print(scores.mean()*(-1))
```

```python
scores = cross_val_score(lr,simulated[["X","X2","X3","X4"]],simulated["Y"],cv=loo,scoring='neg_mean_squared_error')
print(scores.mean()*(-1))
```

```python
#8d -> using a different seed
np.random.seed(67)
X=np.random.normal(0,1,100)
Y=np.random.normal(0,1,100)
Y=X-2*np.power(X,2)+np.random.normal(0,1,100)

sns.scatterplot(x=X,y=Y)
plt.show()
```

```python
simulated=pd.DataFrame({"X": X, "X2": np.power(X,2), "X3":  np.power(X,3), "X4": np.power(X,4), "Y": Y})
simulated.head()
```

```python
scores = cross_val_score(lr,simulated["X"].to_numpy().reshape(-1,1),simulated["Y"],cv=loo,scoring='neg_mean_squared_error')
print(scores.mean()*(-1))
```

```python
scores = cross_val_score(lr,simulated[["X","X2"]],simulated["Y"],cv=loo,scoring='neg_mean_squared_error')
print(scores.mean()*(-1))
```

```python
scores = cross_val_score(lr,simulated[["X","X2","X3"]],simulated["Y"],cv=loo,scoring='neg_mean_squared_error')
print(scores.mean()*(-1))
```

```python
scores = cross_val_score(lr,simulated[["X","X2","X3","X4"]],simulated["Y"],cv=loo,scoring='neg_mean_squared_error')
print(scores.mean()*(-1))
```

As we can see that changing the random seed did not affect the MSE a lot. In fact, the general trend amongst the different models is maintained.<br>

$Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \epsilon$ has the lowest LOOCV error. This because the true function is quadratic.
